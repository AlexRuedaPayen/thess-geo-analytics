## A. Library installation 

First let's install the right libraries

````python
import geopandas as gpd
import matplotlib.pyplot as plt
from thess_geo_analytics.utils.RepoPaths import RepoPaths
from thess_geo_analytics.pipelines.ExtractGeography import ExtractGeography
from pathlib import Path
````

## B. Generate Thessaloniki's Geography

### C. Plotting Thessaloniki

Let's first import Thessaloniki as GeoPandas

````python
gdf = gpd.read_file(aoi_path)
````

Let's then use plotlib to generate a white-bounded plot of Thessaloniki

````python
ax = gdf.plot(edgecolor="black", facecolor="none")
ax.set_title("AOI — Thessaloniki (EL522, NUTS3)")

out = Path("../outputs/figures/aoi_map.png")
out.parent.mkdir(parents=True, exist_ok=True)

plt.savefig(out, dpi=200, bbox_inches="tight")
plt.show()

````
